```python
import yfinance as yf
import pandas as pd

# Download stock data
df = yf.download('TATASTEEL.NS', start='2015-01-01', end='2024-12-31')
df = df[['Open', 'High', 'Low', 'Close', 'Volume']]
df.to_csv('tatasteel_stock_data.csv')
df.head()
```

```python
#Step 2: Visualizing Closing Price from the data
#A line plot helps understand stock trends over the years.


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Close Price')
plt.title("TATASTEEL Closing Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Price (INR)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Step 3: ARIMA Model
#Used for trend-based forecasting. Parameters (5,1,0) were selected manually.



from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df['Close'], order=(5, 1, 0))
arima_result = model.fit()
arima_forecast = arima_result.forecast(steps=30)
arima_forecast.plot(title="ARIMA Forecast")

```

```python
#Step 3(a): SARIMA Model for handling Seasonality for complex processing


from statsmodels.tsa.statespace.sarimax import SARIMAX

# SARIMA(p, d, q)(P, D, Q, s)
# Example seasonal_order: (1, 1, 1, 12) assumes yearly seasonality in monthly data
sarima_model = SARIMAX(df['Close'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_result = sarima_model.fit(disp=False)

# Forecasting next 30 days
sarima_forecast = sarima_result.forecast(steps=30)

# Plotting the SARIMA forecast
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(df['Close'][-100:], label="Actual")
plt.plot(pd.date_range(df.index[-1], periods=30, freq='B'), sarima_forecast, label="SARIMA Forecast")
plt.title("SARIMA Forecast")
plt.legend()
plt.show()
```

```python
#Step 4: Prophet Model
#Meta's Prophet handles trend and seasonality well.

from prophet import Prophet

df_prophet = df.reset_index()[['Date', 'Close']]
df_prophet.columns = ['ds', 'y']
model_prophet = Prophet()
model_prophet.fit(df_prophet)
future = model_prophet.make_future_dataframe(periods=30)
forecast = model_prophet.predict(future)
model_prophet.plot(forecast)
```

```python
#Step 5: LSTM Model
#A deep learning approach to capture complex patterns in stock data (introduces non linearity).



import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df[['Close']])

X, y = [], []
for i in range(60, len(scaled_data)):
    X.append(scaled_data[i-60:i, 0])
    y.append(scaled_data[i, 0])

X, y = np.array(X), np.array(y)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

model_lstm = Sequential()
model_lstm.add(LSTM(50, return_sequences=True, input_shape=(X.shape[1], 1)))
model_lstm.add(LSTM(50))
model_lstm.add(Dense(1))
model_lstm.compile(optimizer='adam', loss='mean_squared_error')
model_lstm.fit(X, y, epochs=5, batch_size=32)
```

```python
#Visualizing the ACTUAL vs PREDICTED LSTM from our Dataset 

predicted = model_lstm.predict(X)
plt.figure(figsize=(10,6))
plt.plot(y, label='Actual')
plt.plot(predicted, label='Predicted')
plt.legend()
plt.title("LSTM Prediction vs Actual")
plt.show()

```

```python
###Step 6: Model Evaluation
#RMSE scores were calculated for all three models to compare performance.


from sklearn.metrics import mean_squared_error
import numpy as np

mse_arima = mean_squared_error(df['Close'][-30:], arima_forecast)
print(f"ARIMA RMSE: {np.sqrt(mse_arima)}")
```

```python
#RMSE calculation for LSTM
predicted_lstm = model_lstm.predict(X)
predicted_lstm = scaler.inverse_transform(predicted_lstm)
actual_lstm = scaler.inverse_transform(y.reshape(-1, 1))

rmse_lstm = np.sqrt(mean_squared_error(actual_lstm, predicted_lstm))
print(f"LSTM RMSE: {rmse_lstm:.4f}")
```
